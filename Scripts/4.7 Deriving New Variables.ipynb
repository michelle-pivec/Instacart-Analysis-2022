### 2. Include two busiest days in busy_day. Include two least busy days in busy_day.

```python
# Create busy_days column 
result = []

for value in merged["order_day"]:
  if value <=1:
    result.append("Busiest day")
  elif value >= 4:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

### 4. Create column for busiest hour of day for orders.

```python
# Identify busiest hours of the day
merged['order_hour'].value_counts(dropna=False)
```

##### Most: 9 - 16; Average: 7, 8, 17-22; Fewest: 0 - 6, 23

```python
# Create busy_hours column by way of for-loop
result = []

for value in merged["order_hour"]:
  if value >=9 and value<=16:
    result.append("Most orders")
  elif value <=6 and value>=0:
    result.append("Fewest orders")
  elif value==23:
    result.append("Fewest orders")
  else:
    result.append("Average orders")
```

```python
# Merge result into df
merged['busy_hours']=result
```

```python
# 5 Check value counts
merged['busy_hours'].value_counts(dropna=False)
```
